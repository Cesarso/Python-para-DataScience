## Criando Novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)

```

```python
# como saber quanto pagar de aluguel. Somando valor + condominio + iptu:
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
#nova variaável criada: Valor Bruto
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
# diminuir as casas decimais

dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)

dados.head(10)
```

```python
# Agora pela vaariavel Tipo, identificar o que é CAsa ou APTO em uma nova variavel
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila', 'Quitinete']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
#apply permite que aplique uma função a cada linha
```

```python
dados
```
